```python
from sklearn import linear_model
import pandas as pd
import datetime
```

```python
data=pd.read_csv('../random_final_final.csv')
data
```

```python
data['Date']=data['Date'].apply(lambda x:datetime.datetime.strptime(x,"%m/%d/%Y"))
data.pop('Percentage Consumed')
#data.pop('Number of Plates Consumed')
X=data
```

```python
X
```

```python
categ=pd.get_dummies(X['Day'])
k=list(categ.columns.values)
print(k)
for i in range(0, len(k)):
    X[k[i]]=categ[k[i]]
```

```python
categ_item=pd.get_dummies(X['Item'])
k=list(categ_item.columns.values)
print(k)
for i in range(0,len(k)):
    X[k[i]]=categ_item[k[i]]
```

```python
X.pop('Item')
X.pop('Day')
X
```

```python
categmonth=pd.get_dummies(X['Month'])
k=list(categmonth.columns.values)
for i in range(0, len(k)):
    X[k[i]]=categmonth[k[i]]
X.pop('Month')
X
```

```python
X['Week']=X['Week'].apply(lambda x: "Week "+str(x))
categweek=pd.get_dummies(X['Week'])
k=list(categweek.columns.values)
for i in range(0, len(k)):
    X[k[i]]=categweek[k[i]]
X.pop('Week')
X
```
